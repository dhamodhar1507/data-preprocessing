Import libraries & Create dataset

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age': [25, np.nan, 35, np.nan, np.nan, 40],
    'city': ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', 'Mumbai'],
    'gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Female'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})

df
```

Fill Missing Values

```python
# Fill missing age with median
df['age'].fillna(df['age'].median(), inplace=True)

# Fill missing city with mode
df['city'].fillna(df['city'].mode()[0], inplace=True)

df
```

Remove duplicates

```python
df.drop_duplicates(inplace=True)
df
```

Encode categorical columns

```python
from sklearn.preprocessing import LabelEncoder

# Label encode gender
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])

# One-hot encode city
df = pd.get_dummies(df, columns=['city'])

df
```

**Min - Max Scaling**

Min-Max Scaling is used when data has no outliers and we need values in a fixed range (0 to 1). It is useful in distance-based algorithms like KNN and Neural Networks.


```python
from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler()

df_minmax = df.copy()
df_minmax[['age', 'annual_income']] = minmax.fit_transform(
    df_minmax[['age', 'annual_income']]
)

df_minmax
```

**Standardisation**

Standardisation is used when data contains outliers or follows normal distribution. It centers the data to mean 0 and variance 1 and is preferred in Logistic Regression, SVM, and PCA.

```python
from sklearn.preprocessing import StandardScaler

standard = StandardScaler()

df_standard = df.copy()
df_standard[['age', 'annual_income']] = standard.fit_transform(
    df_standard[['age', 'annual_income']]
)

df_standard
```
